```python
# ── Cell: すべての画像を image_000001.jpg 形式にリネーム ──
from pathlib import Path
import shutil

ROOT = Path("/content/drive/MyDrive/AffectNet")   # ★ 必要ならここだけ修正 ★
splits = ["Train", "Test"]                         # 走査する 2 つのフォルダ
valid_ext = {".jpg", ".jpeg", ".png"}

for split in splits:
    for expr_dir in (ROOT / split).iterdir():      # anger / happy / …
        if not expr_dir.is_dir():                 # safety
            continue

        imgs = sorted([p for p in expr_dir.iterdir() if p.suffix.lower() in valid_ext])
        if not imgs:
            print(f"[{split}/{expr_dir.name}] 画像が見つかりません")
            continue

        # バックアップ用に元ファイル名を記録する（任意）
        (expr_dir / "_rename_backup.txt").write_text(
            "\n".join([p.name for p in imgs]), encoding="utf-8"
        )

        for idx, p in enumerate(imgs, start=1):
            new_name = f"image_{idx:06d}{p.suffix.lower()}"
            new_path = p.with_name(new_name)
            # 既に同名があれば衝突を避けるため一時名に移動 → 最後に上書き
            if new_path.exists():
                tmp = p.with_name(f"tmp_{idx:06d}{p.suffix.lower()}")
                p.rename(tmp)
                tmp.rename(new_path)
            else:
                p.rename(new_path)

        print(f"[{split}/{expr_dir.name}] → {len(imgs):,} 枚リネーム完了")
```
